```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
x = np.linspace(0,5,11)
y = x**2
```

## plotting methods

1. Functional method
2. Object oriented method

## Functional Method

```python
plt.plot(x,y)
# red color plt.plot(x,y, 'r-')

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Multiplots

plt.subplot(1,2,1) #(num of rows , num of columns, plot number)
plt.plot(x,y,'r')

plt.subplot(1,2,1)
plt.plot(y,x,'b')


plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

## Object Oriented Method

```python
# Recommended

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #4 arguments : [left , bottom , width , height] between 0 to 1
axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')


 
```

```python
# Multiplots

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x,y) 
axes1.set_title(' Lrger One')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y,x) 
axes2.set_title(' Smaller One')
```

## Subplots()

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
# axes is a list of matplotlib axes
for current_ax in axes:
    current_ax.plot(x,y)
    
# same as below
# axes[0].plot(x,y)
# axes[1].plot(x,y)
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2), dpi = 200) 

# figsize is in inches
# dpi : dots per inch
```

## Review 2 methods

```python
# 1

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# 2
fig , axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Adding legends , Labels, Titles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title('Title')
ax.legend(loc=0)
```
